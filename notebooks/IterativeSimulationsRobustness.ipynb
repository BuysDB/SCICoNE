```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns;
from collections import Counter
import glob
from tqdm import tqdm_notebook as tqdm
%matplotlib inline
sns.set(rc={'figure.figsize':(9.7,4.27)})
```

```python
output_path = "/cluster/work/bewi/members/tuncel/sc-dna/notebooks/manuscript_simulations_max_tree_reproducibility/"
all_paths = glob.glob("/cluster/work/bewi/members/tuncel/data/dna/manuscript_/iterative_simulations/*")
scores_df = pd.DataFrame()
for path in tqdm(sorted(all_paths)): # 27
    #import ipdb; ipdb.set_trace()
    dups_per_sim = []
    for sim_id in range(0,100): # 100
        trees_in_folder_in_sim = glob.glob(path+'/'+str(sim_id)+'_*_tree_inferred.txt')
        tree_scores = [] 
        for tree_file in trees_in_folder_in_sim: # 10
            tree_score = np.genfromtxt(tree_file)[0][2]
            tree_scores.append(tree_score)
        max_ts = max(tree_scores)
        max_tree_scores = [ts for ts in tree_scores if ts==max_ts]
        n_dups = len(max_tree_scores) - len(set(max_tree_scores))
        dups_per_sim.append(n_dups)

    scores_df[path.split('/')[-1]] = dups_per_sim
scores_df.to_csv(output_path+"determining_n_iters_1024_coeff" + '.csv')
```
