```python
import pandas as pd
import numpy as np
import glob
import os
import re
from tqdm import tqdm
```

```python
all_folders = glob.glob("../../data/dna/recomb_simulations_25-10-2018/*")
```

```python
n_rep = 100
```

```python
list(range(0,n_rep))
```

#### Get the fixed parameters from the folder name

```python
row_list = []
for folder in tqdm(all_folders):
    fixed_params = re.findall('\d+', folder.split('/')[-1:][0])
    # parse the fixed params from folder name
    n_nodes = fixed_params[0]
    n_regions = fixed_params[1]
    n_reads = fixed_params[2]
    for rep_id in range(0,n_rep):
        inferred_tree = glob.glob(folder+"/"+str(rep_id)+ "_tree_inferred.txt")[0]
        with open(inferred_tree) as f:
            first_line = f.readline()
        inferred_tree_score = float(first_line.split(':')[-1])
        original_tree = glob.glob(folder+"/"+str(rep_id)+ "_tree.txt")[0]
        with open(original_tree) as f:
            first_line = f.readline()
        original_tree_score = float(first_line.split(':')[-1])
        # frobenius norm
        gt = glob.glob(folder+"/"+str(rep_id)+ "_ground_truth.txt")[0]
        inferred_cnvs = glob.glob(folder+"/"+str(rep_id)+ "_inferred_cnvs.txt")[0]
        gt_mat = np.loadtxt(gt)
        inferred_mat = np.loadtxt(inferred_cnvs)
        assert(gt_mat.shape==inferred_mat.shape)
        delta = ((gt_mat - inferred_mat) ** 2).mean()
        regions_known = True
        tup = (rep_id, n_nodes, n_regions, n_reads,delta,original_tree_score, inferred_tree_score, regions_known)
        row_list.append(tup)
```

```python
df = pd.DataFrame(data=row_list, columns=['rep_id','n_nodes','n_regions','n_reads', 'delta', 'likelihood', 'inferred_likelihood', 'regions_known'])
```

```python
df.to_csv("benchmark_segmentation_results.csv")
```
